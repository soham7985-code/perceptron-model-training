```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import Perceptron
from mlxtend.plotting import plot_decision_regions
```

```python
# Load Dataset
from google.colab import files
import io

# Upload the file
print("Please upload your 'placement.csv' file:")
uploaded = files.upload()

# Check if the file was uploaded and read it
if uploaded:
    uploaded_filename = list(uploaded.keys())[0]
    # Directly use the uploaded_filename as it will contain the correct name (e.g., placement.csv, placement (1).csv, etc.)
    try:
        df = pd.read_csv(io.BytesIO(uploaded[uploaded_filename]))
        print("Dataset Shape:", df.shape)
        print("First 5 rows of the dataset:")
        print(df.head())
    except Exception as e:
        print(f"Error reading file '{uploaded_filename}': {e}")
        print("Please ensure you uploaded a valid CSV file with the expected format.")
        df = None
else:
    print("Error: No file was uploaded. Please ensure you select and upload 'placement.csv'.")
    df = None
```

```python
# Visualize Dataset
sns.scatterplot(x=df['cgpa'], y=df['resume_score'], hue=df['placed'])
plt.title("Placement Dataset")
plt.show()
```

```python
# Separate Inputs and Output
X = df.iloc[:,0:2]
y = df.iloc[:,-1]

print(X.head())
print(y.head())
```

```python
# Train Perceptron Model
p = Perceptron()

p.fit(X,y)

print("Model Training Completed")
```

```python
# Coefficients
print("Weights:", p.coef_)
print("Bias:", p.intercept_)
```

```python
# Plot Decision Boundary
plot_decision_regions(X.values, y.values, clf=p, legend=2)
plt.title("Decision Boundary using Perceptron")
plt.show()
```
